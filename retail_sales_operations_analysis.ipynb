```python
#Search Product
def search(table, id):
  try:
      if id in table:
          print(tabulate([table[id]], headers=table[0], tablefmt='fancy_grid'))
      else:
          print("ID does not exist.\n")
  except ValueError:
      print("Invalid input. Please enter a valid number.\n")
```

```python
def display(table):
  if len(table) > 1:
      table_list = list(table.values())
      body = table_list[1:]
      print(tabulate(body, headers=table[0], tablefmt='fancy_grid'))
  else:
      print("No data to display.")
```

Task 5: Saving tables and function to select

```python
def save_table(table, filename):
  table_list = list(table.values())
  with open(f"{filename}.csv", 'w', newline='') as file:
      writer = csv.writer(file)
      writer.writerows(table_list)
  print(f"Table saved to {filename}.csv")

def load_table(filename):
  table = {}
  try:
      with open(f"{filename}.csv", newline='') as file:
          reader = csv.reader(file)
          for idx, row in enumerate(reader):
              if idx == 0:
                  table[0] = row
              else:
                  table[int(row[0])] = row
      print(f"Table loaded from {filename}.csv")
  except FileNotFoundError:
      print(f"No existing file found for {filename}.csv. Starting with an empty table.")
      if filename == 'Product':
        table[0] = ['id', 'name', 'category','price']
      elif filename == 'Customer':
        table[0] = ['id', 'name', 'address','phone']
      elif filename == 'Supplier':
        table[0] = ['id', 'name', 'address','phone']
      elif filename == 'Employee':
        table[0] = ['id', 'name', 'address','phone']
      elif filename == 'Sales Transaction':
        table[0] = ['id', 'customer_id', 'employee_id', 'product_id', 'quantity', 'date_of_sales']
  return table
```

```python
# @title
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
product_df = pd.read_csv('/content/drive/MyDrive/BAIM645/Project_Phase_2/Product.csv')
customer_df = pd.read_csv('/content/drive/MyDrive/BAIM645/Project_Phase_2/Customer.csv')
supplier_df = pd.read_csv('/content/drive/MyDrive/BAIM645/Project_Phase_2/Supplier.csv')
employee_df = pd.read_csv('/content/drive/MyDrive/BAIM645/Project_Phase_2/Employee.csv')
sales_transaction_df = pd.read_csv('/content/drive/MyDrive/BAIM645/Project_Phase_2/Sales Transaction.csv')

# Rename columns of product_df to match expected names
# Assuming the order is id, name, category, price
product_df.columns = ['id', 'name', 'category', 'price']

# ------------------- 1. Distribution of product prices -------------------
plt.figure(figsize=(10, 6))
sns.histplot(product_df['price'], kde=False, bins=20)
plt.title("Distribution of Product Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

print("Price statistics:")
print(product_df['price'].describe())
```

```python
# ------------------- 2. Total quantity sold for each product -------------------
product_df = product_df.rename(columns={'id': 'product_id'})
merged_sales = pd.merge(sales_transaction_df, product_df, how='left', on='product_id')
product_sales = merged_sales.groupby('name')['quantity'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='name', y='quantity', data=product_sales.sort_values(by='quantity', ascending=False).head(10))
plt.title("Top 10 Products by Quantity Sold")
plt.xlabel("Product")
plt.ylabel("Total Quantity Sold")
plt.xticks(rotation=45)
plt.show()
```

```python
# ------------------- 3. Which employee has made the most sales? -------------------
employee_df = employee_df.rename(columns={'id': 'employee_id'})
merged_employee_sales = pd.merge(sales_transaction_df, employee_df, how='left', on='employee_id')
employee_sales = merged_employee_sales.groupby('name')['quantity'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='name', y='quantity', data=employee_sales.sort_values(by='quantity', ascending=False).head(10))
plt.title("Top 10 Employees by Sales Quantity")
plt.xlabel("Employee")
plt.ylabel("Total Quantity Sold")
plt.xticks(rotation=45)
plt.show()
```

```python
# ------------------- 4. Average quantity sold per transaction -------------------
avg_quantity = sales_transaction_df['quantity'].mean()
print(f"Average Quantity Sold per Transaction: {avg_quantity}")
plt.figure(figsize=(10, 6))
sns.histplot(sales_transaction_df['quantity'], kde=False, bins=10)
plt.title("Distribution of Quantity Sold per Transaction")
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.show()
```

```python
# ------------------- 5. Sales by Day of the Week -------------------
sales_transaction_df['date_of_sales'] = pd.to_datetime(sales_transaction_df['date_of_sales'])
sales_transaction_df['day_of_week'] = sales_transaction_df['date_of_sales'].dt.day_name()
sales_by_day = sales_transaction_df.groupby('day_of_week')['quantity'].sum().reset_index()
ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sales_by_day['day_of_week'] = pd.Categorical(sales_by_day['day_of_week'], categories=ordered_days, ordered=True)
sales_by_day = sales_by_day.sort_values('day_of_week')
plt.figure(figsize=(10, 6))
sns.barplot(x='day_of_week', y='quantity', data=sales_by_day)
plt.title("Sales by Day of the Week")
plt.xlabel("Day of Week")
plt.ylabel("Total Quantity Sold")
plt.show()
```

```python
# ------------------- 6. Average price per transaction -------------------
merged_sales_price = merged_sales
merged_sales_price['avg_price'] = merged_sales_price['price']
plt.figure(figsize=(10, 6))
sns.histplot(merged_sales_price['avg_price'], kde=False, bins=20)
plt.title("Distribution of Average Price per Transaction")
plt.xlabel("Average Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# ------------------- 7. Correlation between product price and quantity sold -------------------
plt.figure(figsize=(10, 6))
sns.scatterplot(x='price', y='quantity', data=merged_sales)
plt.title("Correlation between Price and Quantity Sold")
plt.xlabel("Product Price")
plt.ylabel("Quantity Sold")
plt.show()
correlation = merged_sales[['price', 'quantity']].corr()
print(f"Correlation between price and quantity sold: \n{correlation}")
```
